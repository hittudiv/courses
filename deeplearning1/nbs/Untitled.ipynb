```python
import numpy as np
from keras.layers import Dense
from keras.models import Sequential
```

```python
x=np.random.random((100,2))*10
m=[5, 8]
c=3
y=np.dot(x, m)+c
```

```python
model = Sequential([Dense(1, input_shape=(2,))])
model.compile(optimizer="sgd",loss="mse")
```

```python
model.fit(x,y,nb_epoch=100)
```

```python
model.get_weights()
```

```python
model.predict([[2,3],[5,8]])
```
